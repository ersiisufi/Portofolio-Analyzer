```python
import pandas as pd
from src.preprocessing import DataPreprocessor
from src.utils import setup_logging

```

```python
logger = setup_logging()
logger.info("Starting data cleaning process...")
```

```python
# 1. Load the raw price data we saved from Phase 2
raw_prices = pd.read_csv("data/raw/raw_portfolio.csv", index_col="Date", parse_dates=True)
raw_prices.sort_index(inplace=True)
# raw_prices.head
```

```python
# 2. Instantiate our preprocessor object from Phase 3
preprocessor = DataPreprocessor()

# 3. Run the cleaning engine (handles missing values)
cleaned_prices = preprocessor.clean_prices(raw_prices, )

# 4. Calculate both simple and log returns
simple_returns, log_returns = preprocessor.calculate_returns(cleaned_prices)
```

```python
# 5. PERSIST THE DATA: Save to the processed directory for future phases
simple_returns.to_csv("data/processed/simple_returns.csv")
log_returns.to_csv("data/processed/log_returns.csv")
cleaned_prices.to_csv("data/processed/cleaned_prices.csv")

logging.info("Phase 3 complete! Processed return data successfully saved to disk.")
```
