```python
# 1. Load the raw price data we saved from Phase 2
raw_prices = pd.read_csv(RAW_DATA_DIR / "raw_portfolio.csv", index_col="Date", parse_dates=True)
raw_prices.sort_index(inplace=True)
# raw_prices.head
```
