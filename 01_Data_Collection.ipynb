```python
import os
from src.utils import setup_logging
from src.data_loader import FinancialDataLoader
```

```python
# 1. Initialize the shared logging configuration
# This creates the '../logs/portfolio_analyzer.log' file automatically
logger = setup_logging()
logger.info("Starting Phase 1: Data Collection Pipeline.")
```

```python
# 2. Define our investment sandbox parameters
# We handle the portfolio components and the market benchmark separately
portfolio_tickers = ['AAPL', 'MSFT', 'GLD', 'IEF']
benchmark_ticker = ['SPY']
```

```python
start_date = '2016-01-01'
end_date = '2026-01-01'

# 3. Instantiate the Data Loader (pointing to our raw data directory)
loader = FinancialDataLoader(data_dir="./data/raw")
```

```python
# 4. Fetch and archive the Portfolio Assets data
logger.info("Fetching asset portfolio history from Yahoo Finance...")
raw_portfolio_df = loader.fetch_historical_data(
    tickers=portfolio_tickers, 
    start_date=start_date, 
    end_date = end_date,
)
# print(raw_portfolio_df.columns)
loader.save_to_csv(raw_portfolio_df, "raw_portfolio.csv")
```

```python
# 5. Fetch and archive the Benchmark Index data
logger.info("Fetching benchmark index history from Yahoo Finance...")
raw_benchmark_df = loader.fetch_historical_data(
    tickers=benchmark_ticker, 
    start_date=start_date, 
    end_date=end_date
)

loader.save_to_csv(raw_benchmark_df, "raw_benchmark.csv")

logger.info("Phase 2 pipeline execution complete. Datasets successfully hard-saved.")
```
